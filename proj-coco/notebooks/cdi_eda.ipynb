```python
import io
import bson                       # this is installed with the pymongo package
import matplotlib.pyplot as plt
from pathlib import Path
import matplotlib.image as mpimg
from skimage.io import imread
from PIL import Image
# from skimage.data import imread   # or, whatever image library you prefer
import multiprocessing as mp      # will come in handy due to the size of the data

# Constants
dataPath = '/home/dsmith/Documents/analytics-projects/independent-learning/data/coco/raw/cdiscount/'
outPath = '/home/dsmith/Documents/analytics-projects/independent-learning/data/coco/processed/cdiscount/'
```

```python
# Read the bson file into an iterator
data = bson.decode_file_iter(open(dataPath + 'train_example.bson', 'rb'))

# define the objects
prod_id = []
cat_id = []

for ix, d in enumerate(data):
    if ix == 10000:
        break
    # get the product id and category id from the objects
    product_id = d['_id']
    category_id = d['category_id'] 
    for e, pic in enumerate(d['imgs']):
        picture = imread(io.BytesIO(pic['picture']))
        # convert the nd.array to a PIL object
        img = Image.fromarray(picture)
        
        # create the directory if it doesnt exist
        imgDir = outPath + 'category_id=%i' % category_id
        Path(imgDir).mkdir(parents=True, exist_ok=True)
        
        # save the object to a file
        img.save(imgDir + '/%i_%i.jpg' % (product_id, e))
```
